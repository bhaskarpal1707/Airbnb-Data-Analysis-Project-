```python
# Import Liberies:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Data in Notebook:
df = pd.read_csv("Airbnb_US_Raw_Dataset.csv", low_memory=False)
```

```python
df.head()
```

```python
# See the Column Names:
df.columns
```

# Check Missing Values:

```python
print(df.isnull().sum())
```

# Handiling Missing Values:

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month': 0, 'last review' : df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns= ["license", "house_rules"], errors = 'ignore')
```

```python
df.head()
```

```python
# Remove Dollar signs and convert to Float:
df['price'] = df['price'].replace('[\$,]', '', regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex = True).astype(float)
```

```python
df.head()
```

# Remove Duplicates:

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# Descriptive Statistics:

```python
df.describe()
```

# Visualization:

# 1. What is the distribution of listing prices?

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins = 50, kde = True, color = 'red')
plt.title("Distribution of Listing Price")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
â¬† Insights: The histogram shows a fairly even distribution of
listing prices across different price ranges, indicating no
particular concentration of listings in any specific price range.
The KDE line helps visualize this even spread more clearly,
confirming that the dataset contains listings with a wide variety of prices.
```

# 2. How are room type distributed?

```python
df['room type']
```

# 3. How are listings distributed across different neighborhoods?

```python
plt.figure(figsize = (10,7))
sns.countplot(y = "neighbourhood group", data = df, color = "red", order=df['neighbourhood group'].value_counts().index)
plt.title("Number of Listing by Neighbourhood Group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood Group")
plt.show()
```

```python
â¬† Insights: Top most Number of listing by neighbourhood group is 'Manhattan', 
then 'Brooklyn', then 'Queens', then 'Bronx', then 'Staten island' and atlast 'Brookln'. 
```

# 4. hat is relationship between price and room type?

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1')  # Assign to ax
handles, labels = ax.get_legend_handles_labels()  # Get legend handles and labels
plt.legend(handles, labels, title='Room Type')  # Use handles and labels explicitly
plt.title("Price vs Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price ($)")
plt.show()
```

```python
â¬† Insights: Prices of Privite room, Entire home/apt and Shared room are quite similler like the rence of 360 $ to 900 $. 
But hotel room prices range are littlebit diffrent like  nearabout 400 $ to 920 $.
```

# 5. How Has the number of reviews change over with time?

```python
df['last review'] = pd.to_datetime(df['last review'])
review_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize = (12,6))
review_over_time.plot(kind='line', color='red')
plt.title("Number of Reviews Over Time")
plt.xlabel("Date")
plt.ylabel("Numbers of Reviews")
plt.show()
```

```python
â¬† Insights: In year 2014 to 2019 the number of reviews are growing slowly, but in year of 2019 to
2024 the number of reviews aregrowing exponentially. 
```
